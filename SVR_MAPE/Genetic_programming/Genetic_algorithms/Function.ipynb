```python
class CustomSC:
    
    @staticmethod
    def cos(x1, num):
        return np.cos(2*np.pi*(x1/num))
    
    @staticmethod
    def sin(x1, num):
        return np.sin(2*np.pi*(x1/num))
    
    @staticmethod
    def cos_7(x1):return CustomSC.cos(x1, 7)

    @staticmethod
    def sin_7(x1):return CustomSC.sin(x1, 7)
    
    @staticmethod
    def cos_365(x1):return CustomSC.cos(x1, 365)

    @staticmethod
    def sin_365(x1):return CustomSC.sin(x1, 365)
    
    @staticmethod
    def abscos_14(x1):return np.abs(CustomSC.cos(x1, 14))

    @staticmethod
    def abssin_14(x1):return np.abs(CustomSC.sin(x1, 14))
```

```python
class Triangulars:
    """
    - wheather:
        warm1
        cold1
        cold2
        warm2
    - Season:
        winter1
        spring
        summer
        autumn1
        autumn2
        winter2    
"""

# diffuse variables function
    @staticmethod
    def diff(x1, height):
        # Se crea x
        ma = pd.DataFrame();
        x = x1.index.dayofyear.to_list()
        # Se crean variables a, b y c
        ba = pd.DataFrame(x, columns=["x"])
        ba["a"] = height[0]
        ba["b"] = height[1]
        ba["c"] = height[2]
        # Variable difusa
        ca = pd.DataFrame()
        ca["bo"] = (ba.x-ba.a)/(ba.b-ba.a)
        ca["ba"] = (ba.c-ba.x)/(ba.c-ba.b)
        ca = ca.min(axis = 1)
        ca[ca<0]=0
        return ca.to_numpy() 

# wheather var
    @staticmethod
    def warm1(x1):
        hei = [9, 151, 365.25+9]
        return Triangulars.diff(x1, hei)
    
    @staticmethod
    def cold1(x1):
        hei = [151-365.25, 9, 151]
        return Triangulars.diff(x1, hei)
    
    @staticmethod
    def cold2(x1):
        hei = [151, 365.25+9, 365.25+365.25]
        return Triangulars.diff(x1, hei)
    
    @staticmethod
    def warm2(x1):
        hei = [9-365.25, 151-365.25, 9]
        return Triangulars.diff(x1, hei)


# seasonal var
    @staticmethod
    def winter1(x1):
        hei = [310.5-365.25, 8.6529, 138.941176470588]
        return Triangulars.diff(x1, hei)

    @staticmethod
    def spring(x1):
        hei = [8.6529, 138.941176470588, 221.9375]
        return Triangulars.diff(x1, hei)
    
    @staticmethod
    def summer(x1):
        hei = [138.941176470588, 221.9375, 310.5]
        return Triangulars.diff(x1, hei)
    
    @staticmethod
    def autumn1(x1):
        hei = [221.9375, 310.5, 365.25+8.6529]
        return Triangulars.diff(x1, hei)
    
    @staticmethod
    def autumn2(x1):
        hei = [221.9375-365.25, 310.5-365.25, 8.6529]
        return Triangulars.diff(x1, hei)
    
    @staticmethod
    def winter2(x1):
        hei = [310.5, 365.25+8.6529, 365.25+138.941176470588]
        return Triangulars.diff(x1, hei)
```
