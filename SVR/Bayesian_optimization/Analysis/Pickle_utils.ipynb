```python
class pickleUtilss():
    """
    open_pickle: open all specified files
    
    join_pickle: join and ordering sequencial pickles from same file
    
    concat_pred: join prediction of jointed-ordered pickle from all files
    
    concat_error: join error from all files

    """
    
    
    @staticmethod
    def open_pickle(path, files):
        """
        Open Pickles wich contain a dir of dirs
        
        Arguents: 
            path: path where pickle files are located
            files: list of names of files which will be added to the directory
             
        dir_parent: contain all runs dictionary. Labeled by date and time
        dir_son: contain a dictionary of one run. labeled by week . 
        0 contain hyperparameter information.
        """
        import pickle
        import json
        tame = {}
        for ld in files:
            with open(f"{path}\\{ld}", 'rb') as f:
                tame[ld] = pickle.load(f)
        return tame

    @staticmethod
    def join_pickle(pkls):
        """
        Joins dir_sons sequentially by date. Dir labeled by date.
        
        # argument: dir of dirs. 
        # return: dir of ordered prediction.

        """
        li_n = {}
        for a in pkls.keys():
            li_n[a] = pd.DataFrame()
            for i in range(1, len(pkls[a])):
                li_n[a] = pd.concat([li_n[a], pkls[a][i]], axis = 0)
            li_n[a] = li_n[a].sort_index()
        return li_n
    
    @staticmethod
    def concat_pred(j_p):
        """
        Concatenate real value and prediction for all runs.
        
        # argument: dir of jointed dataframes. 
        # return: real values and all predictions 
        
        """
        pred_tot = pd.DataFrame()
        pred_tot["real"] = j_p[[*j_p.keys()][0]]['real']
        name = ['real']
        for a in j_p.keys():
            pred_tot = pd.concat([pred_tot, j_p[a]['predict']], axis = 1)
            name.append(a)
        pred_tot.columns = name
        return pred_tot
    
    @staticmethod
    def concat_error(j_p):
        """
        Concatenate error for all runs.
        
        # argument: dir of jointed dataframes. 
        # return: error given by abs((y - pred)/y)
        
        """
        error_tot = pd.DataFrame()
        for a in j_p.keys():
            error_tot = pd.concat([error_tot, j_p[a]['error']], axis = 1)
        error_tot.columns = [*j_p.keys()]
        return error_tot
```
